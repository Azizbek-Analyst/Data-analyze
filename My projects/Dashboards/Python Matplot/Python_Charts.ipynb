# Stacked Area

```python
stacked_area = pd.read_csv('stacked_area_chart_data.csv')
```

```python
stacked_area.head()
```

```python
colors = ["#011638" , "#7e2987", "#ef2026"]
labels = stacked_area.columns[1:]
sns.set_style("white")
plt.figure(figsize=(15,8))
plt.stackplot(stacked_area['Year'],
              stacked_area['Gas'],
              stacked_area['Petrol'],
              stacked_area['Diesel'],
              colors = colors,
              edgecolor ='none') #remove white lines
plt.xticks(stacked_area['Year'], rotation=45)
plt.legend(labels = labels, loc = "upper left")
plt.ylabel("Number of Cars", fontsize=13)
plt.title("Popularity of Engine Fuel Types (1982-2016)", fontsize=14, weight= "bold")
sns.despine() #bottom=True, right, left, top
plt.show()
```

# Line


```python
line_df = pd.read_csv('line_chart_data.csv')
```

```python
line_df.head()
```

```python
line_df['New Date'] = pd.to_datetime(line_df['Date'])
```

```python
line_df.info()
```

```python
line_df_2008 = line_df[(line_df['New Date'] >= '2008-07-01') & (line_df['New Date'] <= '2008-12-31')]
```

```python
colors = ["#011638" , "#ef2026"]
labels = line_df_2008.columns[1:-1]

plt.figure(figsize=(20,8))

plt.plot(line_df_2008['New Date'], 
         line_df_2008['GSPC500'] , color= colors[1])
plt.plot(line_df_2008['New Date'], 
         line_df_2008['FTSE100'])
plt.title("S&P vs FTSE Returns", fontsize=14, fontweight='bold', color=colors[0])
plt.xlabel('Date')
plt.ylabel('Returns')
plt.legend(labels = labels, loc = "upper right")
plt.show()
```

# Scatter plot

```python
df_scatter = pd.read_csv('scatter_data.csv')
```

```python
df_scatter.head(5)
```

```python
plt.figure(figsize =(12,8))
scatter= plt.scatter(x = df_scatter['Area (ft.)'],y= df_scatter['Price'], 
            alpha=0.6,
            c = df_scatter['Building Type'],
            cmap = 'viridis')

plt.legend(*scatter.legend_elements(),
           loc='upper left',
           title='Building type',
)
plt.xlabel('Area')
plt.ylabel("Price (000'$)")
plt.title('Relationsip between Area and Price of California Real Estate', fontsize=14, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize =(12,8))
sns.scatterplot(y= df_scatter['Price'],
                x = df_scatter['Area (ft.)'],
                hue= df_scatter['Building Type'],
                palette = ['black', 'darkblue','purple', 'pink','white'],
                s =100)
plt.xlabel('Area')
plt.ylabel("Price (000'$)")
plt.title('Relationsip between Area and Price of California Real Estate', fontsize=14, fontweight='bold')
plt.show()
```

# Regression


```python
df_reg = pd.read_csv('scatter_plot_ii.csv')
```

```python
df_reg.head()
```

```python
#plt.figure(figsize=(10,5))
sns.set(rc={'figure.figsize':(10,5)})
sns.regplot(x = 'Budget',
            y = 'Sales',
            data = df_reg,
            scatter_kws = {'color':'k'},
            line_kws = {'color': 'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's $)")
plt.title("Effect of Ad Expenditure on Sales", fontsize=14, fontweight='bold')

plt.show()
```

```python
#plt.figure(figsize=(10,5))
sns.lmplot(x = 'Budget',
            y = 'Sales',
            data = df_reg,
            scatter_kws = {'color':'k'},
            line_kws = {'color': 'red'},
            height=10)
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's $)")
plt.title("Effect of Ad Expenditure on Sales", fontsize=14, fontweight='bold')

plt.show()
```

# Bar and Line chart

```python
bar_line = pd.read_csv('bar_line_chart_data.csv')
```

```python
bar_line.head(5)
```

```python
sns.set_style('white')
fig, ax = plt.subplots(figsize=(10,7))

ax.bar(bar_line['Year'],
       bar_line['Participants'],
       color = 'k')
ax.set_ylabel("Number of Participant", weight='bold')
ax.tick_params(axis = "y",
               width = 2,
               labelsize='large')

ax1 = ax.twinx()
ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax= 1.0))

ax1.plot(bar_line['Year'],
       bar_line['Python Users'],
       color = '#b60000',
       marker= 'D')
ax1.set_ylabel("Python Users", weight='bold', color='#b60000')
ax1.tick_params(axis = "y",
               width = 2,
               labelsize='large',
               colors ='#b60000')
plt.title("KD Nuggets survey Python Users", fontsize = '14', weight='bold')
plt.show()
```
